```python
import matplotlib.pyplot as plt

def alternating_harmonic_sum(n):
    total_sum = 0
    for k in range(1, n + 1):
        total_sum += (-1)**(k - 1) * (1 / k)
    return total_sum

def fib(n):
    a = [1,1]
    if n == 1 or n == 0:
        return a[n]
    for i in range(n):
        a.append(a[-1]+a[-2])
    return a[n]

def example(n):
    total_sum = 0
    for k in range(1, n+1):
        total_sum += (1/2)**k
    return total_sum

n_values = list(range(1, 15))  # Compute for n from 1 to 99

sums = []
for i in n_values:
    sums.append(example(i))

# Equivalent code:
# sums = [alternating_harmonic_sum(n) for n in n_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, sums, marker='o', linestyle='-', color='b')
plt.title('Alternating Harmonic Series Sum')
plt.xlabel('n')
plt.ylabel('Sum')
plt.grid(True)
plt.show()

print(sums[-1])
```

```python
cube = 27
epsilon = 0.01
num_guesses = 0
low = 0
high = cube
guess = (high + low)/2.0
sequence = []
while abs(guess**3 - cube) >= epsilon:
    sequence.append(guess)
    if guess**3 < cube :
        low = guess
    else:
        high = guess
    guess = (high + low)/2.0
    num_guesses += 1
    
sequence.append(guess)
print ('num_guesses =', num_guesses)
print (guess, 'is close to the cube root of', cube)

import matplotlib.pyplot as plt

n_values = list(range(num_guesses+1))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence, marker='o', linestyle='-', color='b')
plt.title('Alternating Harmonic Series Sum')
plt.xlabel('n')
plt.ylabel('Sum')
plt.grid(True)
plt.show()
```

```python
cube = 27
epsilon = 0.01
guess = 0.0
increment = 0.1
num_guesses = 0
sequence = [0]
while abs(guess**3 - cube) >= epsilon:
    guess += increment
    num_guesses += 1
    sequence.append(guess)
print('num_guesses =', num_guesses)
if abs(guess**3 - cube) >= epsilon:
    print('Failed on cube root of', cube)
else:
    print(guess, 'is close to the cube root of', cube)

import matplotlib.pyplot as plt
num_values = num_guesses

n_values = list(range(num_values))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence[:num_values], marker='o', linestyle='-', color='b')
plt.title('Alternating Harmonic Series Sum')
plt.xlabel('n')
plt.ylabel('Sum')
plt.grid(True)
plt.show()
```

# Graphing Functions with Matplotlib

```python

import matplotlib.pyplot as plt

# Generate a range of x values
x_values = [i for i in range(-10, 11)]
# Calculate the corresponding y values
y_values = [x**2 for x in x_values]

# Create the plot
plt.plot(x_values, y_values, label='$y = x^2$')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of $y = x^2$')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```

```python
import math
import matplotlib.pyplot as plt

def sine_taylor(x, n_terms):
    result = 0
    for n in range(n_terms):
        term = ((-1) ** n * x ** (2 * n + 1)) / math.factorial(2 * n + 1)
        result += term
    return result

# Generate x values
x = [i * 0.1 for i in range(-63, 64)]  # x values from -2pi to 2pi with step 0.1

y_sin = [math.sin(xi) for xi in x]

plt.figure(figsize=(10, 6))

# Plot the actual sine function
plt.plot(x, y_sin, label='sin(x)', color='black')

# Plot Taylor series approximations
for n_terms in range(1, 10, 2):  # Plot odd number of terms from 1 to 9
    y_taylor = [sine_taylor(xi, n_terms) for xi in x]
    plt.plot(x, y_taylor, label=f'{n_terms}-term approximation')

plt.title('Taylor Series Approximations of sin(x) at x=0')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(0, color='gray', linewidth=0.5)
plt.axvline(0, color='gray', linewidth=0.5)
plt.legend()
plt.grid(True)

# Set y-axis limits
plt.ylim(-5, 5)

# Show the plot
plt.show()
```
