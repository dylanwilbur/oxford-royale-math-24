```python
import matplotlib.pyplot as plt

def alternating_harmonic_sum(n):
    total_sum = 0
    for k in range(1, n + 1):
        total_sum += (-1)**(k - 1) * (1 / k)
    return total_sum

def fib(n):
    a = [1,1]
    if n == 1 or n == 0:
        return a[n]
    for i in range(n):
        a.append(a[-1]+a[-2])
    return a[n]

n_values = list(range(1, 100))  # Compute for n from 1 to 99

sums = []
for i in n_values:
    sums.append(alternating_harmonic_sum(i))

# Equivalent code:
# sums = [alternating_harmonic_sum(n) for n in n_values]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, sums, marker='o', linestyle='-', color='b')
plt.title('Alternating Harmonic Series Sum')
plt.xlabel('n')
plt.ylabel('Sum')
plt.grid(True)
plt.show()

print(sums[-1])
```

```python
cube = 27
epsilon = 0.01
num_guesses = 0
low = 0
high = cube
guess = (high + low)/2.0
sequence = []
while abs(guess**3 - cube) >= epsilon:
    sequence.append(guess)
    if guess**3 < cube :
        low = guess
    else:
        high = guess
    guess = (high + low)/2.0
    num_guesses += 1
    
sequence.append(guess)
print ('num_guesses =', num_guesses)
print (guess, 'is close to the cube root of', cube)

import matplotlib.pyplot as plt

n_values = list(range(num_guesses+1))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence, marker='o', linestyle='-', color='b')
plt.title('Alternating Harmonic Series Sum')
plt.xlabel('n')
plt.ylabel('Sum')
plt.grid(True)
plt.show()
```

```python
cube = 27
epsilon = 0.01
guess = 0.0
increment = 0.1
num_guesses = 0
sequence = [0]
while abs(guess**3 - cube) >= epsilon:
    guess += increment
    num_guesses += 1
    sequence.append(guess)
print('num_guesses =', num_guesses)
if abs(guess**3 - cube) >= epsilon:
    print('Failed on cube root of', cube)
else:
    print(guess, 'is close to the cube root of', cube)

import matplotlib.pyplot as plt
num_values = num_guesses

n_values = list(range(num_values))

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, sequence[:num_values], marker='o', linestyle='-', color='b')
plt.title('Alternating Harmonic Series Sum')
plt.xlabel('n')
plt.ylabel('Sum')
plt.grid(True)
plt.show()
```

# Graphing Functions with Matplotlib

```python

import matplotlib.pyplot as plt

# Generate a range of x values
x_values = [i for i in range(-10, 11)]
# Calculate the corresponding y values
y_values = [x**2 for x in x_values]

# Create the plot
plt.plot(x_values, y_values, label='$y = x^2$')

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graph of $y = x^2$')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()
```
